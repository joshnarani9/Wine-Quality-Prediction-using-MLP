## wine quality detection using multilayer perceptron model

```python
import pandas as pd
```

```python
data1=pd.read_csv("wine-red.csv")       ### downloaded from uci machine learning repository
```

```python
data2=pd.read_csv("wine-white.csv")
```

```python
data1.head()
```

```python
data2.head()
```

```python
data1.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
X=data1.iloc[:,:-1]
```

```python
X.fillna(X.mean(),inplace=True)
```

```python
Y=data1['quality']
```

```python
Y.fillna(Y.mean(),inplace=True)
```

```python
scaler=StandardScaler()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
scaler.fit(x_train)
```

```python
X_train = scaler.transform(x_train)
X_test = scaler.transform(x_test)
```

## training model

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(512,256,128,64,32,16))
```

```python
mlp.fit(X_train,y_train)
```

```python
pred=mlp.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
print(classification_report(y_test,pred))
```
